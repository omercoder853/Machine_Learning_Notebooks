```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("../Datasets/diamonds.csv")
df
```

```python
df.drop(df[df["x"]==0].index,axis=0,inplace=True)
df.drop(df[df["y"]==0].index,axis=0,inplace=True)
df.drop(df[df["z"]==0].index,axis=0,inplace=True)
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df
```

```python
df.drop(df[df["y"]>20].index,inplace=True,axis=0)
```

```python
df.drop(df[(df["z"]>6) | (df["z"]<2)].index,inplace=True,axis=0)
```

```python
df.drop(df[(df["depth"]>73) | (df["depth"]<52)].index,inplace=True,axis=0)
```

```python
df.drop(df[(df["table"]>71) | (df["table"]<50)].index,inplace=True,axis=0)
```

```python
X = df.drop("price",axis=1)
y = df["price"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
from sklearn.preprocessing import OrdinalEncoder

column_list = ["cut", "clarity", "color"]
cut_list = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
clarity_list = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
color_list = ["J", "I", "H", "G", "F", "E", "D"]
category_lists = [cut_list, clarity_list, color_list]
encoders = dict()
for column, category in zip(column_list, category_lists):
    encoders[column] = OrdinalEncoder(categories=[category])
    X_train[column] = encoders[column].fit_transform(X_train[[column]])
    X_test[column] = encoders[column].transform(X_test[[column]])
```

Modeli kaydederken predict edilecek datanın da hatasız bir şekilde encode ve scale edilmesi için encodersların ve scalerların da kaydedilmesi gerekmektedir.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
model_svr = SVR(C=100,gamma="auto",kernel="rbf")
model_svr.fit(X_train,y_train)
y_preds = model_svr.predict(X_test)
r2 = r2_score(y_test,y_preds)
print(f"R2:%{r2*100:.2f}")
```
