```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df = pd.read_csv("../Datasets/7-cyber_attack_data.csv")
df
```

```python
df.columns
```

```python
X = df.drop("attack_type",axis=1)
y = df["attack_type"]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train,y_train)
y_preds = logistic.predict(X_test)
y_preds
```

```python
acc = accuracy_score(y_test,y_preds)
print(f"Accuracy {acc}")
print(classification_report(y_test,y_preds))
print(f"Confusion Matrix: \n {confusion_matrix(y_test,y_preds)}")
```

```python
from sklearn.multiclass import OneVsOneClassifier,OneVsRestClassifier
```

```python
onevsonemodel = OneVsOneClassifier(LogisticRegression())
onevsrestmodel = OneVsRestClassifier(LogisticRegression())
```

```python
onevsonemodel.fit(X_train,y_train)
y_preds = onevsonemodel.predict(X_test)
acc = accuracy_score(y_test,y_preds)
print(f"Accuracy {acc}")
print(classification_report(y_test,y_preds))
print(f"Confusion Matrix: \n {confusion_matrix(y_test,y_preds)}")
```

```python
onevsrestmodel.fit(X_train,y_train)
y_preds = onevsrestmodel.predict(X_test)
acc = accuracy_score(y_test,y_preds)
print(f"Accuracy {acc}")
print(classification_report(y_test,y_preds))
print(f"Confusion Matrix: \n {confusion_matrix(y_test,y_preds)}")
```
